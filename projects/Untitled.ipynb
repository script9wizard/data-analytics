```python
import pandas as pd

import matplotlib.pyplot as plt

# Load sample dataset
data = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')

# Basic information about the dataset
print("Dataset Info:")
print(data.info())
print("\nFirst 5 rows:")
print(data.head())

# Create figure with subplots
plt.figure(figsize=(15, 10))

# Scatter plot
plt.subplot(2, 2, 1)
plt.scatter(data['sepal_length'], data['sepal_width'])
plt.title('Sepal Length vs Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# Histogram
plt.subplot(2, 2, 2)
plt.hist(data['petal_length'], bins=20, color='green')
plt.title('Petal Length Distribution')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')

# Box plot
plt.subplot(2, 2, 3)
data.boxplot(column=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
plt.title('Feature Distributions')
plt.ylabel('Centimeters')

# Bar plot by species
plt.subplot(2, 2, 4)
data.groupby('species')['petal_width'].mean().plot(kind='bar')
plt.title('Average Petal Width by Species')
plt.ylabel('Petal Width (cm)')

# Adjust layout and show plots
plt.tight_layout()
plt.show()
```
